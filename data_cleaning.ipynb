# Handling the Problem

```python
missing_value = ['N/A', 'na', np.nan]
df = pd.read_csv('work-and-employment-csv-1.csv', na_values=missing_value)
```

```python
df.isnull().sum().sort_values(inplace=True)
```

```python
df.drop(df.columns[[10,11,12,13,14]], axis = 1, inplace=True)
```

```python
df.head()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, xticklabels=False)
```
